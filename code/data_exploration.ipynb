# Data Exploration

## Imports

```python
"""Module providing exploratory data analysis."""

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Data

```python
data = pd.read_csv("../data/datos_modificados.csv")
print(data.shape)
```

```python
data.info()
```

## Begin EDA

```python
round(data.describe().transpose(), 2)
```

## Category Distributions

```python
# ordinal columns
ordinals = ['fragilidad', 'sppb', 'gruposbuena', 'id_3', 'cut_off_points', 'sexo', 'sarcewgsop']

for col in ordinals:
    print(f"{data[col].value_counts()}\n")
```

```python
list(data.cut_off_points.unique())
```

```python
fig, ax = plt.subplots(figsize=(4,3))
bars = data.cut_off_points.value_counts(sort=False)\
            .plot(kind='bar', ax=ax, width=.95, alpha=0.5, rot=0)
ax.bar_label(ax.containers[0])

ax.set_ylabel('Count')
ax.set_xlabel('MQI cut off points')
ax.set_yticks([])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
#plt.show()
fig.tight_layout()
#plt.savefig('../plots/mqi_class_comparison')
```

```python
# Counts in Strength Categories
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10,3))
fig.suptitle("Counts in Strength Categories")

data.sppb.sort_values().value_counts(sort=False).plot(kind='bar', ax=ax1, width=.95, rot=0)
data.fragilidad.sort_values().value_counts(sort=False).plot(kind='bar', ax=ax2, width=.95, rot=0)
data.gruposbuena.sort_values().value_counts(sort=False).plot(kind='bar', ax=ax3, width=.95, rot=0)

ax1.set_xlabel("SPPB; 4=peor, 12=mejor")
ax2.set_xlabel("Fragilidad; 2=peor, 4=mejor")
ax3.set_xlabel("Grupos Buena; 1=peor, 7=mejor")

fig.tight_layout()
plt.show();
```

```python
fix, ax = plt.subplots(figsize=(4,2))
ax.set_title("Counts in Economic Categories")
ax.set_xlabel("ID (3); 1=más rico, 3=más pobre")

data.id_3.sort_values().value_counts(sort=False).plot(kind='bar', ax=ax, width=.95, rot=0)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5), width_ratios=[1, 4])
fig.suptitle("Counts in Other Categories")

data.sexo.sort_values().value_counts(sort=False).plot(kind='bar', ax=ax1, width=.95, rot=0)
data.edad.sort_values().value_counts(sort=False).plot(kind='bar', ax=ax2, width=.95, rot=0)

ax1.set_xlabel("Sexo; 1=Femenina, 2=Masculino")
ax1.bar_label(ax1.containers[0])

ax2.set_xlabel("Edad")
ax2.bar_label(ax2.containers[0])

fig.tight_layout();
```

```python
fig, ax = plt.subplots(figsize=(8,3))
ax.set_title("Comparing Gender Per Age")
ax = pd.crosstab(data.edad,data.sexo).plot(kind='bar', width=.95, rot=45, ax=ax)
ax.set_xlabel("Sexo; 1=Femenina, 2=Masculino");
```

```python
fig, ax = plt.subplots(figsize=(8,4))
fig.suptitle("Handgrip Strength (right)")

data.hand_drch.sort_values().value_counts(sort=False).plot(kind='bar', ax=ax, width=.95, rot=30)

plt.setp(ax.get_xticklabels(), ha="right")
fig.tight_layout();
```

```python
fig, ax = plt.subplots(figsize=(8,3))
ax.set_title("Comparing Handgrip per Gender")
ax = pd.crosstab(data.hand_drch,data.sexo).plot(kind='bar', width=.95, rot=45, ax=ax)
ax.set_xlabel("Sexo; 1=Femenina, 2=Masculino");
```

## Check for Correlations among dependent variables

```python
dependent_variables = ['cut_off_points', 'mqi', 'hand_drch', 'músculo_relativ']
mqi_corr_data = data[dependent_variables].corr(method='spearman')
```

```python
plt.figure(figsize=(4.5,4))
sns.heatmap(mqi_corr_data, annot=True, cmap="coolwarm")
plt.tight_layout();
```

```python
sarco_corr_data = data[['sarcewgsop', 'agil_m_s', 'hand_drch']].corr(method='spearman')
```

```python
plt.figure(figsize=(4,3.5))
sns.heatmap(sarco_corr_data, annot=True, cmap="coolwarm")
plt.tight_layout();
```
